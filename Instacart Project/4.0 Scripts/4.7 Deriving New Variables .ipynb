### Task from 4.7 Exercise 

```python
#creating subset 
df = df_ords_prods_merged [:1000000]
```

```python
df.shape
```

```python
# Define a function 
def price_label(row):
    if row ['prices'] <= 5:
        return 'Low_range_product' 
    elif (row['prices'] > 5)  and (row['prices'] <= 15):
        return 'Mid_range_product'
    elif row ['prices'] > 15: 
        return 'High_range_product' 
    else: return np.nan
```

```python
# Apply the function 

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts (dropna = False) 
```

```python
#checking max product cost in subset
df['prices'].max()
```

```python
#dropping businest day column to create busiest days 
df_ords_prods_merged= df_ords_prods_merged.drop(columns = ['busiest_day'])
```

```python
df_ords_prods_merged.head()
```

```python
# creating busiest days loop
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if (value == 0) or (value ==1):
    result.append("Busiest days")
  elif (value == 4) or (value== 3):
    result.append("Least busiest days")
  else:
    result.append("Regularly busy")
```

```python
#creating new column called busiest day with results output 
df_ords_prods_merged['busiest_days'] = result 
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
# Which hours are the busiest 
df_ords_prods_merged['order_time_of_day'].value_counts(dropna = False)
```

```python
# creating 2nd loop 
result2 = []

for value in df_ords_prods_merged['order_time_of_day']:
    if (value >= 10) and (value <=16):
        result2.append('Most orders')
    elif (value >=0) and (value <=6):
        result2.append('Fewest orders')
    else: result2.append('Average orders')
```

```python
# creating new column for busiest_period_of_day
df_ords_prods_merged['busiest_period_of_day']= result2 
```
